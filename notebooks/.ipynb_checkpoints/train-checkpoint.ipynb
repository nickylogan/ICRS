```python
import numpy as np
import pandas as pd
import sklearn
import sklearn.datasets as ds
import sklearn.model_selection as ms
from sklearn.metrics import mean_squared_error
from math import sqrt
import sklearn.svm as svm
import pickle
```

```python
df_imdd = pd.read_csv('../data/imdd.csv')
df_se = pd.read_csv('../data/se.csv')
df_mi = pd.read_csv('../data/mi.csv')
```

```python
# IMDD
from sklearn.impute import SimpleImputer

imdd_target = df_imdd.GPA
gen_courses = [
"INF0131" ,
"INF0221" ,
"INF0381" ,
"INF0011" ,
"INF0141" ,
"INF0621" ,
"INF0301" ,
"INF0531" ,
"INF0601" ,
"INF0031" ,
"INF0271" ,
"INF0521" ,
"INF0021" ,
"INF0291" ,
]

imdd_feature = df_imdd.filter(items=gen_courses)

# fill missing values with mean column values
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(imdd_feature)
imdd_feature_i = imputer.transform(imdd_feature)

imdd_feature = pd.DataFrame(imdd_feature_i,columns=imdd_feature.columns)

X_train, X_test, y_train, y_test = ms.train_test_split(imdd_feature, imdd_target, test_size=0.3)
X_test
```

```python
clf = svm.SVR(kernel='poly') # Linear Kernel

filename = 'imdd - poly.sav'


clf.fit(X_train, y_train)
pickle.dump(clf,open(filename,'wb'))
y_pred = clf.predict(X_test)
X_t_dF = pd.DataFrame(X_test)

print(clf.score(X_test,y_test))
print(pd.DataFrame({0: y_test, 1: y_pred}))

# print(np.sqrt((y_test-y_pred)*(y_test-y_pred)))
mean_squared_error(y_test, y_pred)
```

```python
# SE

from sklearn.impute import SimpleImputer

se_target = df_se.GPA
se_feature = df_se.filter(items=gen_courses)
# fill missing values with mean column values
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(se_feature)
se_feature_i = imputer.transform(se_feature)

se_feature = pd.DataFrame(se_feature_i,columns=se_feature.columns)

X_train, X_test, y_train, y_test = ms.train_test_split(se_feature, se_target, test_size=0.3)
```

```python
clf = svm.SVR(kernel='poly') # Linear Kernel

filename = 'se - poly.sav'


clf.fit(X_train, y_train)
pickle.dump(clf,open(filename,'wb'))
y_pred = clf.predict(X_test)
X_t_dF = pd.DataFrame(X_test)

print(clf.score(X_test,y_test))
sqrt(mean_squared_error(y_test, y_pred))
# print(np.sqrt((y_test-y_pred)*(y_test-y_pred)))
```

```python
# MI

from sklearn.impute import SimpleImputer

mi_target = df_mi.GPA
mi_feature = df_mi.filter(items=gen_courses)
# fill missing values with mean column values
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(mi_feature)
mi_feature_i = imputer.transform(mi_feature)

mi_feature = pd.DataFrame(mi_feature_i,columns=mi_feature.columns)

X_train, X_test, y_train, y_test = ms.train_test_split(mi_feature, mi_target, test_size=0.3)
```

```python
clf = svm.SVR() # Linear Kernel

filename = 'mi.sav'


clf.fit(X_train, y_train)
pickle.dump(clf,open(filename,'wb'))
y_pred = clf.predict(X_test)
X_t_dF = pd.DataFrame(X_test)

print(clf.score(X_test,y_test))

print(np.sqrt((y_test-y_pred)*(y_test-y_pred)))
```
